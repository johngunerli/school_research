```python
import pandas as pd
# !pip install tabula-py
from tabula import read_pdf

df = read_pdf("./international-schools-in-federal-loan-programs.pdf", pages="all")
# got the file from https://studentaid.gov/sites/default/files/international-schools-in-federal-loan-programs.pdf
```

```python
len(df)
# combine all of these into one dataframe
df = pd.concat(df)
df
```

```python
# drop rows that are just deferment only
df = df[df["Status"] != "Deferment Only"]
# drop opeid and city 
df = df.drop(columns=["OPEID", "City"])
```

```python
# i only want to study specific countries in the world, which are 

countries_of_interest = ["england", "germany", "japan", "the netherlands", "switzerland", "france", "scotland", "canada",]

df = df[df["Country"].str.lower().isin(countries_of_interest)]
```

```python
# upon checking the resulting dataframe, Ox and Cambridge have all of their individual colleges listed, so we need to combine them into one row

# if the row contains "Oxford" or "Cambridge", then just drop them
df = df[~df["School Name"].str.contains("University of Oxford -")]
df = df[~df["School Name"].str.contains("University of Cambridge -")]

# upon checking, there's some unis with certain words that do not have CS programs, so we need to drop them.

some_words = ["Medicine", "Veterinary","Seminary", "Conservatoire", "Hotel", "Chiropractic","Business","Naturopathic", "Dentistry", "Pharmacy", "Optometry", "Podiatry", "Osteopathy", "Theology", "Divinity", "Ministry", "Religion", "Pharmaceutical", "Pharmaceutics", "Pharmacology", "Pharmacognosy", "Pharmaceutical"]

for word in some_words:
    df = df[~df["School Name"].str.contains(word)]
```
